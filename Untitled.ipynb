```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point
```

## Load data

### Data for Concentrations

```python
data_2019 = pd.read_csv('raw_data/Concentration_data/2019/laei_LAEI2019v3_CorNOx15_NO2.csv')
```

```python
data_2016 = pd.read_csv('raw_data/Concentration_data/2016/LAEI2016_2016_NO2.csv')
```

```python
data_2013 = pd.read_csv('raw_data/Concentration_data/2013/PostLAEI2013_2013_NO2.csv')
```

### Data for boroughs

```python
boroughs = gpd.read_file('raw_data/statistical-gis-boundaries-london/ESRI/London_Ward_CityMerged.shp')
```

## Prep data

### Turn to geographic data file

```python
def conc_prep(data, size, year):
    data = data.sample(n=size, random_state=42)
    
    geometry = [Point(xy) for xy in zip(data['x'], data['y'])]
    data_gdf = gpd.GeoDataFrame(data, geometry=geometry)
    
    data_gdf.crs = boroughs.crs
    
    if 'conct' in data_gdf.columns:
        data_gdf.rename(columns={'conct':'conc'}, inplace=True)
    
    data_gdf=data_gdf[['conc','geometry']]
    
    new_conc=f'{year}_data'
    data_gdf.rename(columns={'conc':new_conc}, inplace=True)
    return data_gdf
```

```python
prep_data_2019 = conc_prep(data_2019, 50000, 2019)
```

```python
prep_data_2016 = conc_prep(data_2016, 50000, 2016)
```

```python
prep_data_2013 = conc_prep(data_2013, 50000, 2013)
```

```python
prep_data_2019.head()
```

### Merge with borough

```python
joined_conc_data = pd.concat([prep_data_2019,prep_data_2016,prep_data_2013])
```

```python
joined_data = gpd.sjoin(boroughs, joined_conc_data, how='left')
```

```python
joined_conc_data.head()
```

## Data vis

### Mean concentration per borough

```python
borough_summary = joined_data.groupby('BOROUGH')['2019_data'].mean().reset_index()
borough_summary.columns = ['Borough','Mean Concentration']
print(borough_summary.sort_values('Mean Concentration', ascending = False))
```

```python
def plot_year_data(ax, data, year_column, cmap, norm):
    boroughs.plot(ax=ax, color='white', edgecolor='black')
    data.plot(column=year_column, ax=ax, legend=True, cmap=cmap, markersize=5, norm=norm)
    ax.set_title(f'Air Quality Concentrations in London ({year_column.split("_")[0]})')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 10), sharex=True, sharey=True)
cmap = plt.cm.OrRd
norm = colors.BoundaryNorm(boundaries=np.arange(15, 45, 0.5).tolist() + np.arange(50, 150, 10).tolist(), ncolors=cmap.N, clip=True)
plot_year_data(axs[0], joined_data, '2013_data', cmap, norm)
plot_year_data(axs[1], joined_data, '2016_data', cmap, norm)
plot_year_data(axs[2], joined_data, '2019_data', cmap, norm)
plt.tight_layout()
plt.show()
```
